# Finding the frequency of words in a text

```python
class Text():
    
    def __init__(self):
        
        with open ("text.txt","r",encoding="utf-8") as file:
            
            content_of_text=file.read()
            
            words=content_of_text.split()
            self.pure_words=list()
            
            for i in words:
                i=i.strip("\n")
                i=i.strip(" ")
                i=i.strip(".")
                i=i.strip(",")
                self.pure_words.append(i)
            
    def whole_words(self):  
        
        word_set=set()
        for i in self.pure_words:
            
             word_set.add(i) 
                
        print("****WHOLE WORDS*****")
        
        for i in word_set:
            print(i)
            print("*********")
            
    def word_frequency(self):
        
        word_dictionary=dict()
        for i in self.pure_words:
            if i in word_dictionary:
                word_dictionary[i]+=1
            else:
                word_dictionary[i]=1
                
        
        for word,number in  word_dictionary.items():
            print(" '{}' appears {} times in the text....".format(word,number))
            
            print("******************")
    
            
            
                
            
            
        
            
```

```python
file1=Text()
```

```python
file1.whole_words()
```

```python
file1.word_frequency()
```

# Finding the frequency of letters in a sentence

```python
sentence="ResearchersbelievetheyhavefoundapreviouslyunknownkindofwhaleinwatersoffMexicoswesterncoast"
letter_number=dict()
for i in sentence:
    if i in letter_number:
        letter_number[i]+=1
    else:
        letter_number[i]=1
        
for letter,number in letter_number.items():
    print("'{}' appears {} times in the sentence...".format(letter,number))
    print("*****")

        
```

```python
with open("mail_list.txt","r",encoding="utf-8") as file:
    for line in file:
        line=line.strip("\n")
        if line.endswith(".com") and line.find("@")!=-1:
            print(line)
        
```
