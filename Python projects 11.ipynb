# Imdb  top 250 filmlist and its rating by using request and BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
url="https://www.imdb.com/chart/top"
response=requests.get(url)
html_content=response.content
soup=BeautifulSoup(html_content,"html.parser")


```

```python
film_names=soup.find_all("td",{"class":"titleColumn"})
film_ratings=soup.find_all("td",{"class":"ratingColumn imdbRating"})
```

```python
for film_name,film_rating in zip(film_names,film_ratings):
    a=film_name.text
    b=film_rating.text
    a=a.strip()  
    a=a.replace("\n","")
    a=a.split("     ")[1]
    b=b.strip()  
    b=b.replace("\n","")
    
    
    print("Film name: ",a)
    print("Film rating: ",b)
    print("*******************************")
        
        
```
